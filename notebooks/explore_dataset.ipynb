```python
import pandas as pd
df = pd.read_csv('../data/amazon_data.csv')
df.info()
df.head()
```

```python
#### Cleaning up the dataset by dropping columns that have a lot of NaNs that don't relate to the chatbot:

#selected columns to keep: 
columns_to_keep = [
    'Product Name',
    'Category',
    'Selling Price',
    'Product Specification',
    'Technical Details',
    'About Product',
    'Image',
    'Product Url',
    'Is Amazon Seller'
]

df_clean = df[columns_to_keep].copy()
print("New shape:", df_clean.shape)
df_clean.head()
```

```python
#saving the cleaned dataset
df_clean.to_csv('../data/cleaned_amazon_data.csv', index=False)
```

```python
#seeing what kinds of products dominate the dataset - most common categories

import matplotlib.pyplot as plt

#top 10
top_categories = df_clean['Category'].value_counts().head(10)
top_categories.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Product Categories')
plt.xlabel('Number of Products')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
### Distribution of Selling Prices

import numpy as np

import numpy as np

# Remove $ and commas
price_clean = df_clean['Selling Price'].replace('[\$,]', '', regex=True)

# For ranges like '74.99 - 249.99', split by '-' and take first number
price_clean = price_clean.apply(lambda x: str(x).split('-')[0].strip() if pd.notnull(x) else x)

# Convert to numeric
df_clean['Selling Price Numeric'] = pd.to_numeric(price_clean, errors='coerce')

# Plot histogram with x-axis capped at 1000
df_clean['Selling Price Numeric'].plot(kind='hist', bins=50, figsize=(10, 5), color='orange')
plt.title('Selling Price Distribution (Capped at $1000)')
plt.xlabel('Price ($)')
plt.xlim(0, 1000)  # Limit x-axis range
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Filter to products priced $200 or less
filtered_prices_200 = df_clean[df_clean['Selling Price Numeric'] <= 200]['Selling Price Numeric']

# Calculate stats
median_price = filtered_prices_200.median()
q1 = filtered_prices_200.quantile(0.25)
q3 = filtered_prices_200.quantile(0.75)

# Create summary table
summary_stats = pd.DataFrame({
    "Metric": ["Q1 (25th percentile)", "Median", "Q3 (75th percentile)"],
    "Price ($)": [q1, median_price, q3]
})

# Plot histogram with KDE overlay and quartile lines
plt.figure(figsize=(10,5))
sns.histplot(filtered_prices_200, bins=50, color='orange', kde=True)
plt.axvline(median_price, color='red', linestyle='--', label=f'Median: ${median_price:.2f}')
plt.axvline(q1, color='blue', linestyle=':', label=f'Q1: ${q1:.2f}')
plt.axvline(q3, color='green', linestyle=':', label=f'Q3: ${q3:.2f}')

plt.title('Selling Price Distribution with KDE (Products â‰¤ $200)')
plt.xlabel('Price ($)')
plt.ylabel('Number of Products')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Display the table
summary_stats
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set style
sns.set_theme(style="whitegrid")

# --------------------------------------------------
# 1. Top 10 Product Categories
# --------------------------------------------------
top_categories = df_clean['Category'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(y=top_categories.index, x=top_categories.values, palette="Blues_d")
plt.title('Top 10 Product Categories')
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 2. Price Distribution (â‰¤ $1000)
# --------------------------------------------------
filtered_prices = df_clean[df_clean['Selling Price Numeric'] <= 1000]['Selling Price Numeric']

plt.figure(figsize=(10,5))
sns.histplot(filtered_prices, bins=50, color='orange', kde=True)
plt.title('Selling Price Distribution (Products â‰¤ $1000)')
plt.xlabel('Price ($)')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 3. Price Distribution (â‰¤ $200) with Quartiles
# --------------------------------------------------
filtered_prices_200 = df_clean[df_clean['Selling Price Numeric'] <= 200]['Selling Price Numeric']

median_price = filtered_prices_200.median()
q1 = filtered_prices_200.quantile(0.25)
q3 = filtered_prices_200.quantile(0.75)

summary_stats = pd.DataFrame({
    "Metric": ["Q1 (25th percentile)", "Median", "Q3 (75th percentile)"],
    "Price ($)": [q1, median_price, q3]
})

plt.figure(figsize=(10,5))
sns.histplot(filtered_prices_200, bins=50, color='orange', kde=True)
plt.axvline(median_price, color='red', linestyle='--', label=f'Median: ${median_price:.2f}')
plt.axvline(q1, color='blue', linestyle=':', label=f'Q1: ${q1:.2f}')
plt.axvline(q3, color='green', linestyle=':', label=f'Q3: ${q3:.2f}')
plt.title('Selling Price Distribution with KDE (Products â‰¤ $200)')
plt.xlabel('Price ($)')
plt.ylabel('Number of Products')
plt.legend()
plt.tight_layout()
plt.show()

display(summary_stats)

# --------------------------------------------------
# 4. Price by Category (Top 5 Categories)
# --------------------------------------------------
top5_cats = df_clean['Category'].value_counts().head(5).index
plt.figure(figsize=(10,6))
sns.boxplot(data=df_clean[df_clean['Category'].isin(top5_cats)],
            x='Category', y='Selling Price Numeric')
plt.xticks(rotation=45, ha='right')
plt.title('Price Distribution by Top 5 Categories')
plt.ylabel('Price ($)')
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 5. Missing Data Heatmap
# --------------------------------------------------
plt.figure(figsize=(12,6))
sns.heatmap(df_clean.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 6. Text Length Analysis
# --------------------------------------------------
for col in ['Product Specification', 'Technical Details', 'About Product']:
    df_clean[f'{col}_Length'] = df_clean[col].fillna('').apply(lambda x: len(str(x).split()))

text_lengths = df_clean[['Product Specification_Length', 
                         'Technical Details_Length', 
                         'About Product_Length']].mean().sort_values()

plt.figure(figsize=(8,5))
sns.barplot(x=text_lengths.values, y=text_lengths.index, palette="coolwarm")
plt.title('Average Word Count by Text Field')
plt.xlabel('Average Word Count')
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 7. Image URL Coverage
# --------------------------------------------------
valid_images = df_clean['Image'].notnull().sum()
total_products = len(df_clean)
print(f"Products with an image: {valid_images} / {total_products} ({valid_images/total_products:.2%})")

# --------------------------------------------------
# 8. High Price Outliers (> $1000)
# --------------------------------------------------
high_price_df = df_clean[df_clean['Selling Price Numeric'] > 1000]
print(f"Number of products priced > $1000: {len(high_price_df)}")
display(high_price_df[['Product Name', 'Category', 'Selling Price', 'Product Url']].head(10))
```
